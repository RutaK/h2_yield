```python
#Parameters to change
ro_r = 2.7 
ro_w = 1
phi = 0.0017
Th = 10.7
K = 3.1
U = 2.8


#constants
#Stopping power of minerals (Hoffmann, 1992)
SiA = 1.5
SiB = 1.25
SiG = 1.14
#Yiled of hydrogen per 1 eV (Lin et al., 2005)
GiA = 0.0096
GiB = 0.006
GiG = 0.004

#seconds in ka
time_c = 365*24*60*60*1000
#Gy to eV/kg s conversion
Gy_c = 6.241509e18

# Apparent dosage Gy/ka to eV/kg s
EkA = 0 / time_c * Gy_c
EkB = 0.782/ time_c * Gy_c
EkG = 0.243  / time_c * Gy_c
EthA = 0.061  / time_c * Gy_c
EthB = 0.027  / time_c * Gy_c
EthG = 0.048 / time_c * Gy_c
EuA = 0.218 / time_c * Gy_c
EuB = 0.146 / time_c * Gy_c
EuG = 0.113 / time_c * Gy_c

def calculation(ro_r, ro_w, phi, Th, K, U):

    w = (ro_w*1000 * phi)/ ( ro_r*1000 * (1-phi))
    EiA = (EuA *U * w * SiA)/(1+w*SiA) * GiA + (EthA * Th* w * SiA)/(1+w*SiA) * GiA + (EkA * K * w * SiA)/(1+w*SiA) * GiA
    EiB = (EuB *U * w * SiB)/(1+w*SiB)* GiB + (EthB * Th* w * SiB)/(1+w*SiB)* GiB + (EkB * K * w * SiB)/(1+w*SiB)* GiB
    EiG = (EuG *U * w * SiG)/(1+w*SiG)*GiG + (EthG * Th* w * SiG)/(1+w*SiG)* GiG + (EkG * K * w * SiG)/(1+w*SiG) * GiG
    H2_yield = EiA+EiB+EiG
    
    return H2_yield


result = calculation(ro_r, ro_w, phi, Th, K, U)
print(result)
```
